```python
import torch
from transformers import AutoModelForCausalLM, AutoProcessor, AutoTokenizer
from qwen_vl_utils import process_vision_info
# from dots_ocr.utils import dict_promptmode_to_prompt
```

```python
model_path = "/Users/ljoana/repos/LLMs-Journey/VLMs/DotsOCR"
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    attn_implementation="sdpa",
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)
```

```python



processor = AutoProcessor.from_pretrained(model_path, trust_remote_code=True)

image_path = "demo/demo_image1.jpg"
prompt = """Please output the layout information from the PDF image, including each layout element's bbox, its category, and the corresponding text content within the bbox.

1. Bbox format: [x1, y1, x2, y2]

2. Layout Categories: The possible categories are ['Caption', 'Footnote', 'Formula', 'List-item', 'Page-footer', 'Page-header', 'Picture', 'Section-header', 'Table', 'Text', 'Title'].

3. Text Extraction & Formatting Rules:
    - Picture: For the 'Picture' category, the text field should be omitted.
    - Formula: Format its text as LaTeX.
    - Table: Format its text as HTML.
    - All Others (Text, Title, etc.): Format their text as Markdown.

4. Constraints:
    - The output text must be the original text from the image, with no translation.
    - All layout elements must be sorted according to human reading order.

5. Final Output: The entire output must be a single JSON object.
"""

messages = [
        {
            "role": "user",
            "content": [
                {
                    "type": "image",
                    "image": image_path
                },
                {"type": "text", "text": prompt}
            ]
        }
    ]

# Preparation for inference
text = processor.apply_chat_template(
    messages, 
    tokenize=False, 
    add_generation_prompt=True
)
image_inputs, video_inputs = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    videos=video_inputs,
    padding=True,
    return_tensors="pt",
)

inputs = inputs.to("cuda")

# Inference: Generation of the output
generated_ids = model.generate(**inputs, max_new_tokens=24000)
generated_ids_trimmed = [
    out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
]
output_text = processor.batch_decode(
    generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
)
print(output_text)
```

```python
def display_ocr_results(ocr_data):
    """Display OCR results with proper formatting for different categories"""
    
    for i, element in enumerate(ocr_data):
        category = element['category']
        bbox = element['bbox']
        text = element.get('text', '')
        
        print(f"\n--- Element {i+1} ---")
        print(f"Category: {category}")
        print(f"BBox: {bbox} (x1={bbox[0]}, y1={bbox[1]}, x2={bbox[2]}, y2={bbox[3]})")
        print(f"Dimensions: {bbox[2]-bbox[0]}x{bbox[3]-bbox[1]} pixels")
        
        if text:
            print(f"Text content:")
            if category == "Formula":
                print(f"LaTeX: {text}")
                # Display as Markdown for LaTeX rendering
                display(Markdown(text))
            else:
                print(f"Raw: {text}")
                # Try to render as Markdown
                display(Markdown(text))
        print("-" * 50)

def clean_latex(latex_text):
    """Clean up LaTeX text for better rendering"""
    # Remove extra newlines and clean up formatting
    cleaned = latex_text.replace('\\n', ' ').replace('\\\\\\\\\\n', ' \\\\\\\\ ')
    # Remove leading/trailing $ if present
    if cleaned.startswith('$') and cleaned.endswith('$'):
        cleaned = cleaned[1:-1]
    return cleaned.strip()

def get_formulas_only(ocr_data):
    """Extract only the Formula elements"""
    return [element for element in ocr_data if element['category'] == 'Formula']

def get_text_only(ocr_data):
    """Extract only the Text elements"""
    return [element for element in ocr_data if element['category'] == 'Text']

# Display the results
display_ocr_results(ocr_data)
```

```python
# Advanced LaTeX rendering and analysis functions

def render_formulas_separately(ocr_data):
    """Render only the mathematical formulas with improved LaTeX formatting"""
    formulas = get_formulas_only(ocr_data)
    
    print(f"Found {len(formulas)} mathematical formulas:")
    
    for i, formula in enumerate(formulas):
        print(f"\n=== Formula {i+1} ===")
        print(f"Location: {formula['bbox']}")
        
        # Clean and display the LaTeX
        latex_text = formula['text']
        cleaned_latex = clean_latex(latex_text)
        
        print(f"Original LaTeX:")
        print(latex_text)
        
        print(f"\nCleaned LaTeX:")
        print(cleaned_latex)
        
        # Try to render with Jupyter's LaTeX support
        print(f"\nRendered:")
        try:
            display(Markdown(f"$${cleaned_latex}$$"))
        except:
            display(Markdown(latex_text))

def analyze_layout(ocr_data):
    """Analyze the layout structure of the document"""
    categories = {}
    for element in ocr_data:
        cat = element['category']
        if cat not in categories:
            categories[cat] = []
        categories[cat].append(element)
    
    print("Layout Analysis:")
    print("=" * 30)
    
    for category, elements in categories.items():
        print(f"\n{category}: {len(elements)} elements")
        for i, elem in enumerate(elements):
            bbox = elem['bbox']
            area = (bbox[2] - bbox[0]) * (bbox[3] - bbox[1])
            print(f"  {i+1}. Position: ({bbox[0]}, {bbox[1]}) Size: {bbox[2]-bbox[0]}×{bbox[3]-bbox[1]} Area: {area:,} px²")

def export_to_markdown(ocr_data):
    """Export the OCR results to markdown format"""
    markdown_content = "# OCR Results\n\n"
    
    for i, element in enumerate(ocr_data):
        markdown_content += f"## Element {i+1} ({element['category']})\n\n"
        markdown_content += f"**BBox:** {element['bbox']}\n\n"
        
        if 'text' in element and element['text']:
            if element['category'] == 'Formula':
                # Clean up the LaTeX for markdown
                latex = clean_latex(element['text'])
                markdown_content += f"**LaTeX Formula:**\n\n$${latex}$$\n\n"
            else:
                markdown_content += f"**Text:**\n\n{element['text']}\n\n"
        
        markdown_content += "---\n\n"
    
    return markdown_content

# Run the analysis
analyze_layout(ocr_data)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

# Load the original image
image_path = "page_1.jpg"
img = Image.open(image_path)
img_array = np.array(img)

print(f"Image dimensions: {img.size} (width x height)")
print(f"Image array shape: {img_array.shape}")

# Display the original image
plt.figure(figsize=(12, 8))
plt.imshow(img_array)
plt.axis('off')
plt.title('Original Image')
plt.tight_layout()
plt.show()
```

```python
def draw_bboxes_on_image(img_array, ocr_data, figsize=(15, 10)):
    """Draw bounding boxes on the image with labels"""
    
    # Define colors for different categories
    category_colors = {
        'Formula': 'red',
        'Text': 'blue', 
        'Title': 'green',
        'Caption': 'orange',
        'Table': 'purple',
        'Picture': 'cyan',
        'List-item': 'yellow',
        'Section-header': 'magenta',
        'Page-header': 'brown',
        'Page-footer': 'pink',
        'Footnote': 'gray'
    }
    
    # Create the plot
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    ax.imshow(img_array)
    
    # Draw each bounding box
    for i, element in enumerate(ocr_data):
        bbox = element['bbox']
        category = element['category']
        
        # Extract coordinates (x1, y1, x2, y2)
        x1, y1, x2, y2 = bbox
        width = x2 - x1
        height = y2 - y1
        
        # Get color for this category
        color = category_colors.get(category, 'red')
        
        # Create rectangle patch
        rect = patches.Rectangle(
            (x1, y1), width, height,
            linewidth=2, 
            edgecolor=color, 
            facecolor='none',
            alpha=0.8
        )
        ax.add_patch(rect)
        
        # Add label with element number and category
        label = f"{i+1}: {category}"
        ax.text(x1, y1-10, label, 
               fontsize=10, 
               color=color, 
               fontweight='bold',
               bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    
    ax.set_xlim(0, img_array.shape[1])
    ax.set_ylim(img_array.shape[0], 0)  # Flip y-axis to match image coordinates
    ax.axis('off')
    ax.set_title('OCR Results: Detected Layout Elements', fontsize=16, fontweight='bold')
    
    # Add legend
    legend_elements = []
    for category in set(element['category'] for element in ocr_data):
        color = category_colors.get(category, 'red')
        legend_elements.append(plt.Line2D([0], [0], color=color, lw=2, label=category))
    
    ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1, 1))
    
    plt.tight_layout()
    plt.show()

def show_bbox_details(ocr_data):
    """Print detailed information about each bounding box"""
    print("Bounding Box Details:")
    print("=" * 60)
    
    for i, element in enumerate(ocr_data):
        bbox = element['bbox']
        category = element['category']
        
        x1, y1, x2, y2 = bbox
        width = x2 - x1
        height = y2 - y1
        area = width * height
        
        print(f"\nElement {i+1}: {category}")
        print(f"  Coordinates: ({x1}, {y1}) → ({x2}, {y2})")
        print(f"  Dimensions: {width} × {height} pixels")
        print(f"  Area: {area:,} pixels²")
        
        # Show a preview of the text content
        if 'text' in element and element['text']:
            text_preview = element['text'][:100].replace('\n', ' ')
            if len(element['text']) > 100:
                text_preview += "..."
            print(f"  Text preview: {text_preview}")

# Show bbox details first
show_bbox_details(ocr_data)
```
